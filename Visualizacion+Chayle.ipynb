# Info del Dataset
Este conjunto de datos proporciona información sobre los salarios y atributos de empleo de los ingenieros de datos para el año 2024. Incluye información como salario, título del trabajo, nivel de experiencia, tipo de empleo, residencia del empleado, proporción de trabajo remoto, ubicación de la empresa y tamaño de la empresa. El conjunto de datos permite el análisis de tendencias salariales, patrones de empleo y variaciones geográficas en los roles de ingeniería de datos. Puede ser utilizado por investigadores, analistas y organizaciones para comprender el panorama en evolución del empleo y la compensación en ingeniería de datos.

Archivo a cargar en: https://drive.google.com/drive/folders/1waI3LyDH7xLZVM_Qmqkdp2lz5o08x5MC?usp=sharing

## Descripción de las características:
*	**work_year**: El año en que se recopilaron los datos (2024).
*	**experience_level**: El nivel de experiencia del empleado, categorizado como SE (Ingeniero Senior), MI (Ingeniero de Nivel Medio) o EL (Ingeniero de Nivel Inicial).
*	**employment_type**: El tipo de empleo, como tiempo completo (FT), medio tiempo (PT), contrato (C) o freelance (F).
*	**job_title**: El título o rol del empleado dentro de la empresa, por ejemplo, AI Engineer.
*	**salary**: El salario del empleado en la moneda local (por ejemplo, 202,730 USD).
*	**salary_currency**: La moneda en la que se denomina el salario (por ejemplo, USD).
*	**salary_in_usd**: El salario convertido a dólares estadounidenses para fines de estandarización.
*	**employee_residence**: El país de residencia del empleado.
*	**remote_ratio**: La proporción que indica el grado de trabajo remoto permitido en la posición (0 para sin trabajo remoto, 1 para totalmente remoto).
*	**company_location**: La ubicación de la empresa donde el empleado está empleado.
*	**company_size**: El tamaño de la empresa, a menudo categorizado por el número de empleados (S para pequeña, M para mediana, L para grande).

```python
# Gráfico de caja (boxplot) para mostrar la distribución de salarios según el nivel de experiencia
plt.figure(figsize=(10, 6))
plt.boxplot([df[df['experience_level'] == 'EN']['salary_in_usd'],
             df[df['experience_level'] == 'MI']['salary_in_usd'],
             df[df['experience_level'] == 'SE']['salary_in_usd'],
             df[df['experience_level'] == 'EX']['salary_in_usd']],
            labels=['Entry-Level', 'Mid-Level', 'Senior', 'Expert'],
            patch_artist=True, notch=True, vert=0, boxprops=dict(facecolor='lightblue'))
plt.title('Distribución de salarios según el nivel de experiencia')
plt.xlabel('Salario en USD')
plt.ylabel('Nivel de experiencia')
plt.show()
```

```python
# Gráfico de pastel para mostrar la distribución de tipos de empleo en el campo de la ingeniería de datos
plt.figure(figsize=(8, 6))
df['employment_type'].value_counts().plot(kind='pie', autopct='%1.2f%%', startangle=140, colors=sns.color_palette('Set2'))
plt.title('Distribución de tipos de empleo en el campo de la ingeniería de datos')
plt.ylabel('')
plt.show()

# Nota: Cómo este gráfico no es el mejor para mostrar la totalidad del FullTime, se usa un scatterplot, pero debido al tamaño de sus dimensiones ser muy grande, se utilizará un sample para tomar una muestra.
```

```python
# Gráfico de distribución para mostrar la distribución de tipos de empleo en el campo de la ingeniería de datos

df_distribution = pd.DataFrame(list(df['employment_type'].items()))
df_distribution.columns = ['Tipo de empleo', 'Porcentaje']
plt.figure(figsize=(28, 10))
sns.scatterplot(x='Tipo de empleo', y='Porcentaje', data=df_distribution, palette='Set2', hue="Porcentaje", legend=False)
plt.title('Distribución de tipos de empleo en el campo de la ingeniería de datos')
plt.xlabel('Tipo de empleo')
plt.ylabel('Porcentaje')
plt.show()
```

```python
# Gráfico de barras apiladas para mostrar la relación entre el nivel de experiencia y el tipo de empleo en la ingeniería de datos
seniority_order = ['EN', 'MI', 'SE', 'EX']
employment_by_exp = df.groupby(['experience_level', 'employment_type']).size().unstack()
employment_by_exp.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='viridis')
plt.title('Relación entre el nivel de experiencia y el tipo de empleo en la ingeniería de datos')
plt.xlabel('Nivel de experiencia')
plt.ylabel('Cantidad de empleados')
plt.legend(title='Tipo de empleo')
plt.show()

# En este gráfico se reafirma que al ser la mayoría de los tipo de empleo son FT = Fulltime; no es una variable favorable sobre la cual distribuir
```

```python
# Gráfico de barras horizontales para mostrar la distribución geográfica de las empresas que emplean ingenieros de datos
plt.figure(figsize=(10, 8))
sns.countplot(y='company_location', data=df, palette='colorblind', order=df['company_location'].value_counts().index, hue='company_location', legend=False)
plt.title('Distribución geográfica de las empresas que emplean ingenieros de datos')
plt.xlabel('Cantidad de empresas')
plt.ylabel('Ubicación de la empresa')
plt.show()

# Nota: Debido a que aquí también la mayoría de los casos son de US el gráfico no se visualiza bien, por lo cual se reemplaza por una tabla de frecuencias.
```

```python
relative_frequencies = df['company_location'].value_counts(1).mul(100)
frequency_table = pd.DataFrame({
    'Frecuencia Absoluta': df['company_location'].value_counts(),
    'Frecuencia Relativa': relative_frequencies,
    'Frecuencia Acumulada': relative_frequencies.cumsum()
})
frequency_table.reset_index(inplace=True)
frequency_table.columns = ['Ubicación', 'Frecuencia Absoluta', 'Frecuencia Relativa', 'Frecuencia Acumulada']
frequency_table.head();

df_copy = df.copy()
top = df['company_location'].value_counts().nlargest(10).index
print(top)
df_copy.loc[~df_copy.company_location.isin(top), 'company_location'] = 'Otros'

sns.countplot(y='company_location', data=df_copy, palette='colorblind', order=df_copy['company_location'].value_counts().index, hue='company_location', legend=False)
```

```python
# Gráfico de violín para mostrar la relación entre el nivel de experiencia y el salario en USD
plt.figure(figsize=(10, 6))
sns.violinplot(x='experience_level', y='salary_in_usd', data=df, palette='pastel', hue='experience_level', legend=False)
plt.title('Relación entre el nivel de experiencia y el salario en USD')
plt.xlabel('Nivel de experiencia')
plt.ylabel('Salario en USD')
plt.show()
```

```python
# Gráfico de boxplot para mostrar la distribución de salarios según el tipo de empleo y el nivel de experiencia
plt.figure(figsize=(12, 8))
sns.boxplot(x='employment_type', y='salary_in_usd', hue='experience_level', data=df, palette='Set2')
plt.title('Distribución de salarios según el tipo de empleo y el nivel de experiencia')
plt.xlabel('Tipo de empleo')
plt.ylabel('Salario en USD')
plt.legend(title='Nivel de experiencia', loc='upper left')
plt.show()

# Duda para el tutor: En las condiciones decía de agregar hue para hacer gráficos más ricos, pero en este caso, como vimos, la mayoría del tipo de empleo es full time;
```
